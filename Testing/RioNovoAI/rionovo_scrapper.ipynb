```python
import requests

START_ID_TEXT = '<h1 class="entry-title"><a href="'
END_ID_TEXT = '</a></h1>'
LEN_START = len(START_ID_TEXT)

articles = []
current_page = 1
while True:
  all_text_in_page = requests.get('https://rionovo.wordpress.com/page/'+str(current_page)).text
  if all_text_in_page.find("This is somewhat embarrassing, isn&rsquo;t it?") != -1:
    break
  for article in all_text_in_page.split("<article "):
    if article.startswith("id="):
      start_id = article.find(START_ID_TEXT)
      end_id = article.find(END_ID_TEXT, start_id)
      header_text = article[start_id+LEN_START:end_id]
      post_link = header_text[:header_text.find('"')]
      post_title = header_text[header_text.find('>')+1:]
      articles.append((post_title, post_link))
    if len(articles)%10 == 0:
      print("Articles found so far:", len(articles))
  current_page += 1
```

```python
len(articles)
```

```python
with open("rionovo_articles.txt", "w") as fout:
  for article in articles:
    fout.write(article[0]+","+article[1]+"\n")
```

```python
for id, article in enumerate(articles):
    article_text = requests.get(article[1]).text
    with open("article_"+str(id)+".txt", "w") as fout:
        fout.write(article[1]+'\n')
        fout.write(article[0]+'\n')
        fout.write("--------------\n")
        fout.write(article_text[article_text.find('<div class="entry-content">'):article_text.find("<!-- .entry-content -->")])
    if id%10 == 0:
        print("Articles saved so far:", id)
```
